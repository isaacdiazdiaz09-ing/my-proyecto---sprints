### [C3] Para cada producto, ¿cuál es la proporción de las veces que se pide y que se vuelve a pedir?

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Agrupar por product_id y calcular el total de órdenes y el total de reórdenes para cada producto
product_reorder_stats = order_products.groupby('product_id')['reordered'].agg(['sum', 'count']).reset_index()
product_reorder_stats.rename(columns={'sum': 'total_reorders', 'count': 'total_orders_for_product'}, inplace=True)

# Calcular la tasa de reorden
product_reorder_stats['reorder_rate'] = product_reorder_stats['total_reorders'] / product_reorder_stats['total_orders_for_product']

# Fusionar con el DataFrame 'products'
products_names_df = products[['product_id', 'product_name']]
product_reorder_rates = product_reorder_stats.merge(products_names_df, on='product_id', how='left')

min_orders_threshold = 50
reliable_reorder_rates = product_reorder_rates[product_reorder_rates['total_orders_for_product'] >= min_orders_threshold]

print(f"Top 10 productos con la mayor tasa de reordenación (con al menos {min_orders_threshold} órdenes):")
print(reliable_reorder_rates.sort_values(by='reorder_rate', ascending=False).head(10))

print(f"\nTop 10 productos con la menor tasa de reordenación (con al menos {min_orders_threshold} órdenes):")
print(reliable_reorder_rates.sort_values(by='reorder_rate', ascending=True).head(10))

# Visualización 
plt.figure(figsize=(10, 6))
sns.histplot(reliable_reorder_rates['reorder_rate'], bins=30, kde=True, color='darkblue')
plt.title('Distribución de la Tasa de Reordenación por Producto')
plt.xlabel('Tasa de Reordenación')
plt.ylabel('Número de Productos')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```
