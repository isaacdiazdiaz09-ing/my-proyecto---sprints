Paso 2: Preparar los datos

```python
import pandas as pd
import numpy as np


#Reemplaza los nombres de las columnas 
print("Renombrando columnas a minúsculas...")
df.columns = df.columns.str.lower()
print("Nuevos nombres de columnas:", df.columns.tolist())

# Columna 'user_score': Reemplazar 'tbd' por NaN y convertir a float
print("\nProcesando 'user_score'...")
# Asegurarse de que 'tbd' se maneje correctamente
df['user_score'] = df['user_score'].replace('tbd', np.nan)
df['user_score'] = pd.to_numeric(df['user_score'], errors='coerce') 
print("Tipo de dato de 'user_score' después de la conversión:", df['user_score'].dtype)

# Columnas numéricas (sales, critic_score): ya deberían ser numéricas o convertibles.
df['critic_score'] = pd.to_numeric(df['critic_score'], errors='coerce')
print("Tipo de dato de 'critic_score' después de la conversión:", df['critic_score'].dtype)

# Para 'year_of_release', 'name' y 'genre', los NaNs son cruciales.
# Eliminaremos filas con NaNs en estas columnas.
initial_rows = df.shape[0]
df.dropna(subset=['name', 'year_of_release', 'genre'], inplace=True)
rows_after_dropping_critical_na = df.shape[0]
print(f"\nFilas eliminadas debido a NaN en 'name', 'year_of_release' o 'genre': {initial_rows - rows_after_dropping_critical_na}")


# Ahora que los NaNs críticos han sido manejados, convertir 'year_of_release' a int
df['year_of_release'] = df['year_of_release'].astype(int)
print("Tipo de dato de 'year_of_release' después de la conversión:", df['year_of_release'].dtype)


# Manejo de valores ausentes restantes:

print("\nManejando NaNs en 'rating'...")
df['rating'].fillna('unknown', inplace=True)
print("Valores únicos en 'rating' después de manejar NaNs:", df['rating'].unique())

# La decisión es dejarlos como NaN. Imputar podría sesgar el análisis de su impacto en las ventas.
# La ausencia de una puntuación podría ser una característica en sí misma 
print("\nValores nulos restantes después del preprocesamiento:")
print(df.isnull().sum())

print("\nPrimeras 5 filas después de la preparación de datos:")
print(df.head())
print("\nInformación general del dataset después de la preparación de datos:")
df.info()

# Calcula las ventas totales (la suma de las ventas en todas las regiones)
# y coloca estos valores en una columna separada.
print("\nCalculando ventas totales...")
df['total_sales'] = df['na_sales'] + df['eu_sales'] + df['jp_sales'] + df['other_sales']
print("Columna 'total_sales' creada. Primeros 5 valores:")
print(df['total_sales'].head())

```

Paso 3. Analiza los datos

```python
import pandas as pd

relevant_data_start_year = 2013
current_year = 2016 


df_relevant = df.query('@relevant_data_start_year <= year_of_release <= @current_year').copy()

print(f"Periodo de datos relevante seleccionado: desde {relevant_data_start_year} hasta {current_year}.")
print(f"Número de filas en el DataFrame relevante: {df_relevant.shape[0]}")
print(f"Años únicos en el DataFrame relevante: {df_relevant['year_of_release'].unique()}")

from scipy import stats as st
import numpy as np

alpha = 0.05

xone_sales = df_relevant.loc[df_relevant['platform'] == 'XOne', 'total_sales']
pc_sales = df_relevant.loc[df_relevant['platform'] == 'PC', 'total_sales']

if xone_sales.empty or pc_sales.empty:
    print("Error: Uno o ambos grupos de plataformas están vacíos o no existen en el df_relevant.")
else:
    levene_statistic, p_value_levene = st.levene(xone_sales, pc_sales)

    if p_value_levene > alpha:
        equal_variances = True
    else:
        equal_variances = False

    ttest_statistic, p_value_ttest = st.ttest_ind(xone_sales, pc_sales, equal_var=equal_variances)

    if p_value_ttest < alpha:
        print("Hay una diferencia ESTADÍSTICAMENTE SIGNIFICATIVA en las ventas promedio de juegos entre Xbox One y PC.")
    else:
        print("No hay evidencia estadística suficiente para decir que hay una diferencia significativa en las ventas promedio de juegos entre Xbox One y PC.")
```

```python
import pandas as pd
from scipy import stats as st
import numpy as np
 
# relevant_data_start_year = 2013
# current_year = 2016 # Asumimos que 2016 es el último año completo de datos para planificar 2017



# PASO 5: Prueba de Hipótesis con Prueba de Levene 

# Define tu nivel de significancia (alfa)
alpha = 0.05

print("--- Prueba de Hipótesis: Ventas promedio de juegos en Xbox One vs PC (Período relevante 2013-2016) ---")

# 1. Prepara los datos: Divide las ventas totales por plataforma
#    Aquí tomamos 'total_sales' de 'df_relevant' para las plataformas 'XOne' y 'PC'.
xone_sales = df_relevant.loc[df_relevant['platform'] == 'XOne', 'total_sales']
pc_sales = df_relevant.loc[df_relevant['platform'] == 'PC', 'total_sales']

# Verifica que los grupos no estén vacíos antes de proceder con las pruebas
if xone_sales.empty or pc_sales.empty:
    print("Error: Uno o ambos grupos de plataformas (Xbox One o PC) están vacíos o no tienen datos en el período relevante.")
    print("Asegúrate de que 'XOne' y 'PC' son nombres correctos de plataformas en tu DataFrame y que hay datos para ellas entre 2013 y 2016.")
else:
    # 2. Realiza la Prueba de Levene para la igualdad de varianzas
    #    Hipótesis Nula (H0_Levene): Las varianzas de las ventas de XOne y PC son iguales.
    #    Hipótesis Alternativa (H1_Levene): Las varianzas de las ventas de XOne y PC son diferentes.
    print("\nRealizando la Prueba de Levene para la igualdad de varianzas...")
    levene_statistic, p_value_levene = st.levene(xone_sales, pc_sales)

    print(f"Estadístico de Levene: {levene_statistic:.4f}")
    print(f"Valor p de la Prueba de Levene: {p_value_levene:.4f}")

    # 3. Determina el valor de 'equal_var' basado en el resultado de Levene
    if p_value_levene > alpha:
        equal_variances = True
        print(f"El valor p de Levene ({p_value_levene:.4f}) es mayor que {alpha}.")
        print("No podemos rechazar la hipótesis nula de varianzas iguales. Se utilizará equal_var=True en la prueba t.")
    else:
        equal_variances = False
        print(f"El valor p de Levene ({p_value_levene:.4f}) es menor o igual que {alpha}.")
        print("Rechazamos la hipótesis nula de varianzas iguales. Se utilizará equal_var=False en la prueba t.")

    # 4. Realiza la prueba t de Student para medias de dos poblaciones independientes
    #    Hipótesis Nula (H0_ttest): Las ventas promedio de juegos en Xbox One y PC son iguales.
    #    Hipótesis Alternativa (H1_ttest): Las ventas promedio de juegos en Xbox One y PC son diferentes.
    print("\nRealizando la Prueba t de Student para la comparación de medias de ventas...")
    ttest_statistic, p_value_ttest = st.ttest_ind(xone_sales, pc_sales, equal_var=equal_variances)

    print(f"Estadístico de la Prueba t: {ttest_statistic:.4f}")
    print(f"Valor p de la Prueba t: {p_value_ttest:.4f}")

    # 5. Conclusión de la prueba t
    print("\n--- Conclusión de la Prueba t ---")
    if p_value_ttest < alpha:
        print(f"El valor p ({p_value_ttest:.4f}) es menor que el nivel de significancia ({alpha}).")
        print("Rechazamos la hipótesis nula.")
        print("Conclusión: Hay una diferencia ESTADÍSTICAMENTE SIGNIFICATIVA en las ventas promedio de juegos entre Xbox One y PC.")
    else:
        print(f"El valor p ({p_value_ttest:.4f}) es mayor o igual que el nivel de significancia ({alpha}).")
        print("No podemos rechazar la hipótesis nula.")
        print("Conclusión: No hay evidencia estadística suficiente para decir que hay una diferencia significativa en las ventas promedio de juegos entre Xbox One y PC.")

    # Opcional: Mostrar las medias para ayudar a interpretar la dirección de la diferencia si es significativa
    print(f"\nVentas promedio de juegos para Xbox One: {xone_sales.mean():.2f} millones USD")
    print(f"Ventas promedio de juegos para PC: {pc_sales.mean():.2f} millones USD")
```
