3.2-N√∫mero de rese√±as de usuarios y la calificaci√≥n promedio para cada libro.

```python
query_tarea_2 = """
SELECT 
    book_id,
    COUNT(username) AS num_reviews,
    AVG(rating) AS avg_rating
FROM 
    ratings
GROUP BY 
    book_id
ORDER BY 
    num_reviews DESC, avg_rating DESC;
"""

# Ejecutar la consulta y mostrar el resultado (solo los 5 m√°s revisados)
df_tarea_2 = pd.io.sql.read_sql(query_tarea_2, con=engine)
display(df_tarea_2.head())
```

An√°lisis y Conclusi√≥n:Este resultado revela una desconexi√≥n clave
El libro m√°s popular (ID 948, 160 rese√±as) tiene la calificaci√≥n promedio m√°s baja (3.66) entre el top 5.

3.3-Editorial con el mayor n√∫mero de libros con m√°s de 50 p√°ginas.

```python
query_tarea_3 = """
SELECT 
    t2.publisher,
    COUNT(t1.book_id) AS total_books
FROM 
    books AS t1
INNER JOIN 
    publishers AS t2 ON t1.publisher_id = t2.publisher_id
WHERE 
    t1.num_pages > 50
GROUP BY 
    t2.publisher
ORDER BY 
    total_books DESC
LIMIT 1;
"""

# Ejecutar la consulta y mostrar el resultado
df_tarea_3 = pd.io.sql.read_sql(query_tarea_3, con=engine)
display(df_tarea_3)
```

Conclusi√≥n y Observaci√≥n 
Penguin Books es la editorial m√°s estrat√©gica, al haber publicado 42 libros de lectura sustancial (m√°s de 50 p√°ginas), excluyendo as√≠ folletos o publicaciones cortas. Su enfoque en el contenido de longitud significativa se alinea directamente con el comportamiento del usuario que busca un producto de lectura completo.
Propuesta de Valor: La estrategia de adquisici√≥n de contenido para el nuevo producto debe priorizar y maximizar la relaci√≥n con Penguin Books. Esto garantiza un flujo constante de material de lectura significativo, vital para la retenci√≥n a largo plazo de los usuarios.

3.4-Autor con la m√°s alta calificaci√≥n promedio del libro

```python
query_tarea_4 = """
WITH RatedBooks AS (
    SELECT 
        book_id,
        AVG(rating) AS book_avg_rating
    FROM 
        ratings
    GROUP BY 
        book_id
    HAVING 
        COUNT(rating_id) >= 50
)
SELECT
    t3.author,
    AVG(t1.book_avg_rating) AS author_avg_rating
FROM
    RatedBooks AS t1
INNER JOIN
    books AS t2 ON t1.book_id = t2.book_id
INNER JOIN
    authors AS t3 ON t2.author_id = t3.author_id
GROUP BY
    t3.author
ORDER BY
    author_avg_rating DESC
LIMIT 1;
"""

# Ejecutar la consulta y mostrar el resultado
df_tarea_4 = pd.io.sql.read_sql(query_tarea_4, con=engine)
display(df_tarea_4)
```

Conclusi√≥n y Observaci√≥n:  
El autor/a J.K. Rowling/Mary GrandPr√© ostenta la calificaci√≥n promedio m√°s alta (4.28) entre los libros populares (con al menos 50 ratings). Esto demuestra una calidad excepcional y consistente que resuena con un amplio segmento de usuarios.
Propuesta de Valor: Este autor es el est√°ndar de oro de la plataforma. El nuevo producto deber√≠a basar sus recomendaciones iniciales y su curaci√≥n de contenido en el estilo o g√©nero de este autor para maximizar la satisfacci√≥n del nuevo usuario. Adicionalmente, se puede crear una "Biblioteca de Colecci√≥n Maestra" destacando a este autor.

3.5-N√∫mero promedio de rese√±as de texto entre los usuarios que calificaron m√°s de 50 libros.

```python
query_tarea_5 = """
WITH HighRaters AS (
    SELECT 
        username
    FROM 
        ratings
    GROUP BY 
        username
    HAVING 
        COUNT(book_id) > 50
)
SELECT
    -- Contar el total de rese√±as de texto para esos usuarios (t1.review_id) 
    -- y dividir por el n√∫mero total de usuarios expertos
    CAST(COUNT(t1.review_id) AS REAL) / (SELECT COUNT(username) FROM HighRaters) AS avg_text_reviews_per_user
FROM
    reviews AS t1
INNER JOIN
    HighRaters AS t2 ON t1.username = t2.username;
"""

# Ejecutar la consulta y mostrar el resultado
df_tarea_5 = pd.io.sql.read_sql(query_tarea_5, con=engine)
display(df_tarea_5)
```
